```python
import torch
from tqdm.notebook import tqdm
import matplotlib.pyplot as plt
```

```python
x = torch.tensor(5.).requires_grad_()

opt = torch.optim.Adam([x], lr=.1)

xs = []
step_sizes = []
for i in tqdm(range(50)):
    xs.append(x.detach().clone())

    loss = 0.2*x.abs()
    opt.zero_grad()
    loss.backward()
    opt.step()
    
    step_sizes.append(torch.abs(x.detach().clone()-xs[-1]))
    

plt.plot(xs)
plt.show()
plt.plot(step_sizes)
plt.show()
```
