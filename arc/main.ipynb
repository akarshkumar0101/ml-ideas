```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
```

```python
c = np.array([mcolors.to_rgba(i)[:3] for i in mcolors.TABLEAU_COLORS.keys()])
```

```python
import json
  
f = '/Users/akarshkumar0101/arc/training/3bdb4ada.json'
with open(f) as json_file:
    data = json.load(json_file)
```

```python
for i in data['train']:
    x = c[np.array(i['input'])]
    y = c[np.array(i['output'])]
    plt.subplot(121)
    plt.imshow(x); plt.colorbar()
    plt.subplot(122)
    plt.imshow(y); plt.colorbar()
    plt.show()
```
